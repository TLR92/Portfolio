```python
import csv
import os
os.chdir("/content/drive/MyDrive/Colab Notebooks")
```

```python
# Lettura del file ed inserimento delle righe in una lista
pilots = []
with open("formula1_data.csv") as pilots_file:
  pilots_reader = csv.reader(pilots_file)
  for i, row in enumerate(pilots_reader):
    if i == 0:
      pass
    else:
      pilots.append(row)
# Creazione di un dizionario: Chiave: Costruttore, Valore: Lista piloti
pilots_team = {}
for i in pilots:
  if i[1] in pilots_team:
    if i[0] not in pilots_team[i[1]]:
      pilots_team[i[1]].append(i[0])
    else:
      pass
  else:
    pilots_team[i[1]] = [i[0]]
# Creazione di un dizionario: Chiave: Pilota, Valore: Lista dei piazzamenti
pilots_results = {}
for i in pilots:
  if i[0] in pilots_results:
    pilots_results[i[0]].append(i[4])
  else:
    pilots_results[i[0]] = [i[4]]



pilots_results
```

```python
# Sostituzione dei piazzamenti con i rispettivi punti guadagnati
for d in pilots_results:
  for i, r in enumerate(pilots_results[d]):
    if pilots_results[d][i] == "1":
      pilots_results[d][i] = 10
    elif pilots_results[d][i] == "2":
      pilots_results[d][i] = 8
    elif pilots_results[d][i] == "3":
      pilots_results[d][i] = 6
    elif pilots_results[d][i] == "4":
      pilots_results[d][i] = 5
    elif pilots_results[d][i] == "5":
      pilots_results[d][i] = 4
    elif pilots_results[d][i] == "6":
      pilots_results[d][i] = 3
    elif pilots_results[d][i] == "7":
      pilots_results[d][i] = 2
    elif pilots_results[d][i] == "8":
      pilots_results[d][i] = 1
    else:
      pilots_results[d][i] = 0
pilots_results
```

```python
# Definizione della prima funzione (Lista con punti, vittorie e podi del pilota)
def driver_stats(driver):
  total_points = 0
  total_wins = 0
  total_podiums = 0
  for i in pilots_results[driver]:
    total_points += int(i)
    if i == 10:
      total_wins += 1
    if i > 5:
      total_podiums += 1
  driver_list = [total_points, total_wins, total_podiums]

  return driver_list
```

```python
driver_stats("Kubica")
```

```python
# Definizione della seconda funzione (Creazione file .txt con classifica piloti)
def drivers_ranking():
  drivers_totalpoints = {}
  for i in pilots_results:
    drivers_totalpoints[i]=sum(pilots_results[i])
  ranking_d = dict(sorted(drivers_totalpoints.items(), key=lambda item: item[1], reverse=True))
  with open("Drivers_Standings_2008.txt", "w") as ranking_file:
    ranking_file.write(f"Drivers Standing 2008 Formula 1\n")
    for driver in ranking_d:
      ranking_file.write(f"{driver}: {ranking_d[driver]}\n")
  return ranking_file
```

```python
drivers_ranking()
```

```python
# Definizione della terza funzione (Dizionario con classifica costruttori)
def teams_ranking():
  drivers_standings = {}
  with open("Drivers_Standings_2008.txt") as ranking_file:
    for i, row in enumerate(ranking_file):
      if i == 0:
        pass
      else:
        drivers_standings[row.split(":")[0]] = int(row.split(":")[1])
  teams_standings = {}
  for i in pilots_team:
    points = 0
    for j in pilots_team[i]:
      points += drivers_standings[j]
    teams_standings[i] = points
  ranking_t = dict(sorted(teams_standings.items(), key=lambda item: item[1], reverse=True))
  return ranking_t

```
